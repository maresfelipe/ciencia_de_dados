# Descritivo e exploração de dados com pandas (groupbys)

```python
import pandas as pd
```

```python
alunos = pd.read_csv("alunos.csv")
alunos
```

```python
#criei um novo data frame com apenas as colunas que usarei do data frame maior
colunasSelecionadas = ['sexo', 'av1']
sexoeav1 = alunos.filter(items=colunasSelecionadas)

#retornei o novo data frame
sexoeav1.head()
```

```python
#Quero descobrir qual foi a média de cada sexo:
sexoeav1.groupby('sexo').mean()
```

```python
#Quero ver o histograma das notas de ambos os sexos:
sexoeav1.groupby('sexo').hist()
```

```python
#mostrando estatistica mais completas:
sexoeav1.groupby('sexo').describe()
```
